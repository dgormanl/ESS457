# Magnesite Solubility Example of Iterative Calculations

### Import a module necessary for the calculations

```python
import math as m
```

#### Set up parameters that don't change through the calculcations

```python
# solubility product for magnesite
Ksp=10**-5.05

# Debye-Huckel parameters
A=0.5085
B=0.3281
aMg=5.5
acarb=5.4
```

### The main code for the iterative calculations and printing the results

```python
# initial calulation of copper and sulfate molality
iMg=m.sqrt(Ksp)
icarb=iMg
# initial calculation of ionic strength
Iinit=0.5*(iMg*4+icarb*4)

# headers for the results
print('Iter\t\tI\t\tgamma Mg\tgamma Carb\t\t[Mg]\t\t[carb]')
print('----------------------------------------------------------------------------------------------------------')


# printing of the initial result
print(' %i\t\t %.4f\t\t  %.3f\t\t   %.3f\t\t%.3e\t\t%.3e' % (0,Iinit,1,1,iMg,icarb))

# calculation of activity coefficients
actMg=(-A*2**2*Iinit**0.5)/(1+(B*aMg*Iinit**0.5)) 
actcarb=(-A*2**2*Iinit**0.5)/(1+(B*acarb*Iinit**0.5))

# recalculation of molality from activity coefficients
mMg=m.sqrt(Ksp/(10**actMg*10**actcarb))
mcarb=mMg

# calculation of ionic strength with the new molalities
I=0.5*(mMg*4+mcarb*4)

# printing of the first iteration result
print(' %i\t\t %.4f\t\t  %.3f\t\t   %.3f\t\t%.3e\t\t%.3e' % (1,I,10**actMg,10**actcarb,mMg,mcarb))



# setting up an iterator
iterations=2

# loop for sucessive iterations with termination criteria based on ionic strength
if Iinit != I:
    Inew=Iinit
    while (abs(Inew-I)>0.001 ):
        if Inew == Iinit:
            Inew = I
        else:
            I=Inew
        
        actMg=(-A*4*m.sqrt(I))/(1+(B*aMg*m.sqrt(I)))
        actcarb=(-A*4*m.sqrt(I))/(1+(B*acarb*m.sqrt(I)))
        
        gammaMg=10**actMg
        gammacarb=10**actcarb
        
        mMg=m.sqrt(Ksp/(gammaMg*gammacarb))
        mcarb=mMg

        Inew=0.5*(mMg*4+mcarb*4)

        print(' %i\t\t %.4f\t\t  %.3f\t\t   %.3f\t\t%.3e\t\t%.3e' % (iterations,Inew,gammaMg,gammacarb,mMg,mcarb))
        iterations=1+iterations

print("------------------------------------------Final Results---------------------------------------------------")
print(' %i\t\t %.4f\t\t  %.3f\t\t   %.3f\t\t%.3e\t\t%.3e' % (iterations,Inew,gammaMg,gammacarb,mMg,mcarb))
```
