# Root solver
You can use this notebook to solve for roots of polynomials. This will be helpful for homework 3.

```python
# load the numpy package
import numpy as np
```

# Example
The following cell will solve for the roots of the following equation to determine $H^+$ concentration: $-2*10^{13.09}x^3-x^2+10^{-11.27}$. Note that including zero is necessary in the coefficients since there is no `x` term in the equation. The only reasonable root for this problem is 6.028e-9, which corresponds to $pH = 8.22$.

```python
p = [-2*10**13.09, -1,0, 10**-11.27]
roots=np.roots(p)
for i in range(len(roots)):
    print(roots[i])
```
